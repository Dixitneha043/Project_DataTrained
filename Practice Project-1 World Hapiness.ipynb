# Removing Outliers

```python
print(df)
```

```python
#we checked that thers is outlier in Family Column so we can use IQR to remove outliers
def removeOutlier(col):
    sorted(col)
    qua1,qua2 = col.quantile([0.25,0.75])
#IQR calculate 25 to 75% of data
    IQR= qua2 - qua1
    lowerRange = qua1 - (1.5*IQR)
    upperRange= qua2+(1.5*IQR)
    return lowerRange, upperRange
```

```python
#removeing outliers from Family Column
lowScore,highScore = removeOutlier(df['Family'])
df['Family']=np.where(df['Family']>highScore,highScore,df['Family'])
df['Family']= np.where(df['Family']<lowScore,lowScore,df['Family'])
```

```python
df['Family'].plot.box() #after remove outliers, plotting Family column again
```

Now we can see that there is no Outlier
